```python
import pandas as pd
df = pd.read_csv("/Aicte Intership Dataset.csv")
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
# Categorical/Text columns
text_cols = ["Job Title","Job Type","Company Name","Posted Date","Cities","States",
             "Stipend","Start Date","Duration","Late date to apply"]

for col in text_cols:
    df[col] = df[col].fillna("Unknown")

# Numerical column
df["Numer of Openings"] = df["Numer of Openings"].fillna(df["Numer of Openings"].median())
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# Convert date columns
date_cols = ["Posted Date","Late date to apply"]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], format="%d-%m-%Y", errors="coerce")

# Convert openings to integer
df["Numer of Openings"] = df["Numer of Openings"].astype(int)
```

```python
df.head()
```

```python
df["Stipend_num"] = df["Stipend"].str.extract(r'(\d+)').astype(float)
```

```python
df.head()
```

```python
# Convert openings to integer
#df["Stipend_num"] = df["Stipend_num"].astype(int)
df.drop("Stipend", axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Stipend_num"] = df["Stipend_num"].fillna(df["Stipend_num"].median())
# Categorical/Text columns
text_cols = ["Posted Date","Late date to apply"]

for col in text_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.to_csv("final_cleaned_internships.csv", index=False)
```
